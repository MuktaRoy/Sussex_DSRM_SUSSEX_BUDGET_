```python
import pandas as pd

# Reading in the data
data_csv = pd.read_csv('/movie_metadata.csv')

print(data_csv.head())
print(data_csv.info())


#remove noisy/duplicate/duplicate  data
initial_rows = data_csv.shape[0]
data_csv.drop_duplicates(inplace=True)
final_rows = data_csv.shape[0]
duplicates_removed = initial_rows - final_rows
data_csv.dropna(inplace=True)
print(f"final datasets to use: {duplicates_removed}")

# Display parameter for each column
for column in data_csv.columns:
    unique_values = data_csv[column].unique()
    print(f"Column: {column}")
    print(unique_values)
    print("=" * 50)
data_csv['title_year'] = pd.to_datetime(data_csv['title_year'], format='%Y')
data_csv['profit'] = data_csv['gross'] - data_csv['budget']

# final  dataset
print("\n preprocessed Dataset:")
print(data_csv.info())
data_csv.to_csv('cleaned_movie_data.csv', index=False)
cleaned_movie_data = pd.read_csv('cleaned_movie_data.csv')

# cleaned dataset
print(cleaned_movie_data.head())
print(cleaned_movie_data.info())
```
